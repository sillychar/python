```python
def get_passport(text):
    new_text = text.split('\n')
    
    passport_number = re.search('[A-Z]{1}\d{7}',text)
    date_of_issue = re.search('\d{2}[/]\d{2}[/]\d{4}',text)
    
    for i,line in enumerate(new_text):
        if re.findall('<<<',line):
#             print(i,line)
            current_city = new_text[i-2]
            birth_place = new_text[i-3]
#             print(birth_place)
    passport_attributes = {'Passport Number':[passport_number.group()],'Date of Issue':[date_of_issue.group()],
                          'Current City':[current_city], 'Place of Birth': [birth_place]}
        
    return passport_attributes
#     return date_of_issue.group()

passport_attributes = get_passport(text)
print(passport_attributes)
df = pd.DataFrame(passport_attributes)
df.to_csv('Chetan_passport.csv')
df.to_excel('Chetan_passport.xlsx')
df.to_json('Chetan_passport.json')
df.T
```
